```python
import pandas as pd

data_ridhwan = pd.read_clipboard()

# Menampilkan data
print(data_ridhwan)
```

```python
ridhwan = data_ridhwan[data_ridhwan['Bedrooms'] == 2]

# Menampilkan nama
print(ridhwan)
```

```python
ridhwan['Bathrooms'] = pd.to_numeric(ridhwan['Bathrooms'])

import numpy as np

ridhwan['Bathrooms'] = ridhwan['Bathrooms'].apply(lambda x: 'large' if x > 2 else 'small')

# Menampilkan DataFrame setelah modifikasi
print(ridhwan)
```

```python
import numpy as np

ridhwan['newvariable'] = np.where(ridhwan['Offers'] > 2, 'large', 'small')

# Menampilkan DataFrame 'nama' setelah penambahan kolom baru
print(ridhwan)
```

```python
# Menambahkan kolom baru 'newvariable' 
ridhwan['newvariable'] = ridhwan['Price'] / ridhwan['SqFt']

# Menampilkan DataFrame 'nama' setelah penambahan kolom baru
print(ridhwan)
```

```python
ridhwan = ridhwan.drop(columns=['newvariable'])

# Menampilkan DataFrame 'nama' 
print(ridhwan)
```

```python
kolom1dan2 = data_ridhwan.iloc[:, 0:2]

# Menampilkan DataFrame kolom1dan2
print(kolom1dan2)
```

```python
# Memilih kolom 1 dan 2 dari DataFrame data_nama
kolom3dan4 = data_ridhwan.iloc[:, 2:4]

# Menampilkan DataFrame kolom3dan4
print(kolom3dan4)
```

```python
# Menggabungkan dua DataFrame 
kolom1sd4 = pd.concat([kolom1dan2, kolom3dan4], axis=1)

# Menampilkan DataFrame kolom1sd4
print(kolom1sd4)
```

```python
# Menggabungkan baris dari dua DataFrame
baris1sd3 = data_ridhwan.iloc[0:3, :]
baris4sd6 = data_ridhwan.iloc[3:6, :]
baris1sd6 = pd.concat([baris1sd3, baris4sd6], ignore_index=True)
# Append sudah tidak berlaku di pandas terbaru, jadinya saya ganti -ridhwan
# Menampilkan DataFrame baris1sd6
print(baris1sd6)
```
